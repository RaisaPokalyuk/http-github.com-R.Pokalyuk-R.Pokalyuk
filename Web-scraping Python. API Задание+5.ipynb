```python
#подключаем библиотеки
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
#адрес страницы для парсинга
link="https://tandemtools.ru/catalog/dreli/"
```

```python
#отправка запроса на получение кода страницы
anwers=requests.get(link)
```

```python
#проверка статуса ответа на запрос
anwers.status_code
```

```python
#текст ответа на запрос
anwers.text;
```

```python
#размещаем текст
soup=BeautifulSoup(anwers.text,'lxml')
soup;
```

```python
#в размеченном тексте нужно найти тег div
#со значением параметра class="product-card js-item-card"
#воспользуемся методом find_all
#soup.find_all("div",attrs={"class":"product-card js-item-card"})

drels=soup.find_all("div",
                    attrs={'class':['product-card js-item-card with-discount',
                                    'product-card js-item-card']})

#type(drels)

#здесь мы получили список,в каждом элементе которого содержится информация по
#одной позиции

drels[0];

len(drels)
```

```python
#разберем первую позицию
drels[0].find('a',attrs={'class':'ttl'})
```

```python
#извлекаем название дрели
drels[0].find('a',attrs={'class':'ttl'}).get_text(strip=True)
```

```python
#product-card js-item-card тэг а
#price hot-price тэг span
#product-card js-item-card with-discount тэг a
```

```python
#разобрали 1 позицию

drell_name=drels[0].find('a',attrs={'class':'ttl'}).get_text(strip=True)
price=drels[0].find('span',attrs={'class':'price hot-price'}).get_text(strip=True)
drell_price=int(''.join(price.split()))


print(drell_name,drell_price)
```

```python

#выводим все дрели с одной страницы
dct_drels={'название дрели': None,
           'цена дрели': None,}

lst_drels=[]

for drel in drels:
    dct_drels['название дрели']=drel.find('a',attrs={'class':'ttl'}).get_text(strip=True)
    price=drel.find('span',attrs={'class':['price',
                                           'price hot-price']}).get_text(strip=True)
    dct_drels['цена дрели']=int(''.join(price.split()))
    lst_drels.append(dct_drels.copy())

pd.DataFrame(lst_drels)
# print(drels_name,drels_price)

```

```python
df=pd.DataFrame(columns=["название дрели","цена дрели"]) #создаем пустой DF для вывода столбцов
```

```python
#пагинация -обход всех страниц
link="https://tandemtools.ru/catalog/dreli/?page="
for page in list(range(1,24)):
    dsa=link+str(page)
    # print(dsa)
    anwers=requests.get(dsa) #отправляем запрос
    soup=BeautifulSoup(anwers.text,'lxml')
    drels=soup.find_all('div',
                        attrs={'class':['product-card js-item-card with-discount',
                                    'product-card js-item-card']})
    dct_drels={'название дрели': None,
           'цена дрели': None,}

    lst_drels=[]

    for drel in drels:
        dct_drels['название дрели']=drel.find('a',attrs={'class':'ttl'}).get_text(strip=True)
        price=drel.find('span',attrs={'class':['price',
                                           'price hot-price']}).get_text(strip=True)
        dct_drels['цена дрели']=int(''.join(price.split()))
        lst_drels.append(dct_drels.copy())

    df=pd.concat([df,pd.DataFrame(lst_drels)])
    time.sleep(10)

#     dct_drels={'название дрели': None,
#            'цена дрели': None,}

# lst_drels=[]

# for drel in drels:
#     dct_drels['название дрели']=drel.find('a',attrs={'class':'ttl'}).get_text(strip=True)
#     price=drel.find('span',attrs={'class':'price hot-price'}).get_text(strip=True)
#     dct_drels['цена дрели']=int(''.join(price.split()))
#     lst_drels.append(dct_drels.copy())

# pd.DataFrame(lst_drels)  #разбор текста,который получили


# df = pd.concat([df, pd.DataFrame(lst_drels)], ignore_index=True) #подклеиваем к DF
# time.sleep(10)
```
